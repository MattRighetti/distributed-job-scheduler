```python
import matplotlib.pyplot as plt
import pandas as pd

data1 = pd.read_csv('prova1.csv')
data2 = pd.read_csv('prova2.csv')
data3 = pd.read_csv('prova3.csv')

x1 = data1[data1.columns[0]]
y1 = data1[data1.columns[1]]
x2 = data2[data2.columns[0]]
y2 = data2[data2.columns[1]]
x3 = data3[data3.columns[0]]
y3 = data3[data3.columns[1]]

plt.plot(x1,y1, "r")
plt.plot(x2,y2, "b")
plt.plot(x3,y3, "g")
plt.xticks(pd.np.arange(0, 60, 10))
plt.yticks(pd.np.arange(min(y1), max(y1)+1, 5))
plt.xlabel('TimeStamp')
plt.ylabel('NumJobs')
plt.show()
```
