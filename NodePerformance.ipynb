```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('prova.csv')

x = data[data.columns[0]]
y = data[data.columns[1]]

plt.plot(x,y)
plt.xticks(pd.np.arange(0, 60, 10))
plt.yticks(pd.np.arange(min(y), max(y)+1, 5))
plt.xlabel('TimeStamp')
plt.ylabel('NumJobs')
plt.show()
```
