# Performance analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

# 3 nodes with jobs that vary from 500ms to 2000ms

```python
data1 = pd.read_csv('../StressTests/run-3-500-2000/cluster1.log.csv')
data2 = pd.read_csv('../StressTests/run-3-500-2000/cluster2.log.csv')
data3 = pd.read_csv('../StressTests/run-3-500-2000/cluster3.log.csv')

start = 22
end = 68

x1 = data1[data1.columns[0]][start:end]
y1 = data1[data1.columns[1]][start:end]
x2 = data2[data2.columns[0]][start:end]
y2 = data2[data2.columns[1]][start:end]
x3 = data3[data3.columns[0]][start:end]
y3 = data3[data3.columns[1]][start:end]

figure(num=None, figsize=(15, 15), dpi=80, facecolor='w', edgecolor='k')
plt.plot(x1, y1, "r")
plt.plot(x2, y2, "b")
plt.plot(x3, y3, "y")
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False) # labels along the bottom edge are off
plt.gcf().autofmt_xdate()
plt.xlabel('TimeStamp')
plt.ylabel('NumJobs')
plt.show()
```
